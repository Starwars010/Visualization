```python
#세개의 선을 
#첫번째 빨강 실선
#두번째 그린 점선
#세번째 블루 삼각형


data_1={'data_x': [1,2,3,4], 'data_y':[2, 2.5, 3.3, 4.3]}
data_2={'data_x': [1,2,3,4], 'data_y':[2, 3,   3,   6]}
data_3={'data_x': [1,2,3,4], 'data_y':[2, 3,   5,   10]}

plt.plot('data_x','data_y', data=data_1, label='리플', 'r-')
plt.plot('data_x','data_y', data=data_2, label='이더리움', 'g:')
plt.plot('data_x','data_y', data=data_3, label='비트코인', 'b^')

plt.xlabel('x축', fontdict={'size':12})
plt.ylabel('y축', fontdict={'size':12})

plt.legend(frameon=False) 

plt.show()

#x,y의 개수가 맞아야 함
#모양을 자세하게 주려면 즉 옵션을 다양하게 주려면 딕셔너리 말고 리스트로 할 것!!
```

```python
import numpy as np #계산 관련 라이브러리
a = np.arange(0,2,0.2) #0부터 2미만까지 0.2씩 리스트 생성

plt.plot(a,a, 'r--', linewidth=10) #선의 두께 설정
plt.plot(a,a**2, 'bo', markersize=20) #마커의 크기설정 / 원의 크기 
plt.plot(a,a**3, 'g-.')

# plt.grid(True) #그리드 설정--> x,y축 모두 설정한 거
plt.grid(True, axis='y') #그리드 설정-->y축만 설정
plt.show()
```

```python
import numpy as np #계산 관련 라이브러리
a = np.arange(0,2,0.2) #0부터 2미만까지 0.2씩 리스트 생성

plt.plot(a,a, 'r--', linewidth=10) #선의 두께 설정
plt.plot(a,a**2, 'bo', markersize=20) #마커의 크기설정 / 원의 크기 
plt.plot(a,a**3, 'g-.')

# plt.grid(True) #그리드 설정--> x,y축 모두 설정한 거
plt.grid(True, axis='y') #그리드 설정-->y축만 설정

#plt.title('그래프제목')
plt.title('그래프제목', loc='right', fontdict={'fontsize':16, 'fontweight':'bold'}) #loc default:가운데

plt.show()
```

```python
#막대그래프 bar

z=np.arange(3)

x=['2020','2021','2022']
y=[100, 400, 900]

plt.bar(z,y) #x자리에 z를 넣어야 함
plt.xticks(z,x)  #z의 0,1,2 3개값을 x값으로 치환하겠다는 의미

plt.show()
```

```python
z=np.arange(3)

x=['2020','2021','2022']
y=[100, 400, 900]

#막대그래프 옵션
plt.bar(z,y, width=0.6, align='edge', color='green') 
plt.xticks(z,x)  

plt.show()
```

```python
z=np.arange(3)

x=['2020','2021','2022']
y=[100, 400, 900]

#막대그래프 옵션
plt.barh(z,y) #수평막대그래프
plt.yticks(z,x)  #시간개념이 없게함

plt.show()
```

```python
#산점도 scatter plot

np.random.seed(1) #random.seed : 난수 발생시 일정한 값으로 난수 발생 #ex> 항상 3이 나옴 #한번 나온 값을 고정시킨다고 보면됨
n=50

x=np.random.rand(n) #np.random.rand 은 난수를 발생시키는데 0~1미만 사이의 난수를 발생
y=np.random.rand(n)

plt.scatter(x,y)
plt.show()

#seed가 있기 때문에 재생 버튼 누를 때마다 같은 거 50개 발생
#만약 seed가 없으면 누를 때 마다 새로운 50개의 값을 생성시킴
```

```python
np.random.seed(1) 
n=50

x=np.random.rand(n) 
y=np.random.rand(n)

size=(30*np.random.rand(n))**2     #0~1사이 값이므로 보통 퍼센트값을 구할 떄 많이 씀 / 근데 너무 작으니까 30배 해주고 제곱까지 했음

plt.scatter(x,y, s=size)
plt.show()
```

```python
np.random.seed(1) 
n=50

x=np.random.rand(n) 
y=np.random.rand(n)

size=(30*np.random.rand(n))**2 
color=np.random.rand(n)   

plt.scatter(x,y, s=size, c=color, alpha=0.6) #alpha=투명도
plt.show()
```

```python
np.random.seed(1) 
n=50

x=np.random.rand(n) 
y=np.random.rand(n)

size=(30*np.random.rand(n))**2 
color=np.random.rand(n)   

plt.scatter(x,y, s=size, c=color, alpha=0.4, cmap='Set1') 

plt.colorbar()

plt.show()
```
